So we have 5000 handdrawn digits with 784 pixels values for each. The 1st column is the true classification of the digit

We first look at a single digit's worth of data to make sure we understand what the data represents (the first digit happens to be a 9).

```python
%%R
onedig = as.numeric(train.sub[1,-1])
onedigmat = matrix(onedig, nrow=28)
image(onedigmat)
```

We can see that it is in fact a 9, but that the orientation is not correct. We need to rotate the image in order to have it accurately reflect what was drawn.

By multiplying our data matrix by the correct rotation matrix we are able to get the image to draw correctly.

```python
%%R
rotmat = diag(rep(1, times=28))[28:1,]
newmat = onedigmat%*%rotmat
image(newmat)
```

This will be important when we go to interpret our Principal Components.

We can investigate which pixels seem to vary in our dataset using a heatmap (in particular we are using the heatmap.2 function from the gplots package):

We can see from this that all the ”action” is happening in the center of the images. In particular, many
of the pixels far from the center have virtually no variance (because they are always 0). This suggests
that this dataset is likely to benefit greatly from dimension reduction methods, as those pixels are not
useful in the first place.

We now turn to principle components as our dimension reduction method of choice. We can implement PCA by hand or use one of a number of different existing functions.

We now look to classifying the full dataset.

```python
%%R
allEigs = eigen(var(sampleTrain[, -1]))
pcDatAll = as.matrix(sampleTrain[, -1]) %*% allEigs$vectors[, 1:25]

allPCDF = cbind.data.frame(pcDatAll, label = sampleTrain$label)
```

```python
%%Rcaching --cacherandom
system.time({
    allCVres = cvRpart2(form = label ~ ., dat = allPCDF, truth = sampleTrain$label, 
        folds = 10, cps = seq(5e-06, 0.01, length.out = 10), verbose = FALSE, mc.cores = 3)
})
```

We then create our rpart fit using the optimal cp chosen above:

```python
%%R
alldigsRp = rpart(label ~ ., allPCDF, method = "class", control = rpart.control(cp = allCVres$bestCp))
```

This gives us a training data confusion matrix of 

```python
%%R
table(truth = sampleTrain$label, preds = predict(alldigsRp, allPCDF, type = "class"))
```

We then apply our classifier to the independent test data to get a (more or less) unbiased estimator of the misclassification rate

```python
%%R
allPCTest = as.matrix(test[, -1]) %*% allEigs$vectors[, 1:25]
allTestDF = cbind.data.frame(allPCTest, label = test$label)
alltestPreds = predict(alldigsRp, allTestDF, type = "class")
table(truth = test$label, pred = alltestPreds)
```

The error rate for this is

```python
%%R
sum(alltestPreds != test$label)/nrow(test)
```
